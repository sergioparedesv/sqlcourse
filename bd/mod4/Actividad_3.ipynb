## Ejercicio 3

Define una función llamada is_romano() dentro de la clase Numero que a partir de una cadena de caracteres, devuelve True si esa cadena de caracteres corresponde con un número romano y falso en caso contrario. Después, modifica el método para que lance un error en el caso de que el valor que nos pasan por parámetro no se corresponde con el patrón de un número romano.

```python
# En esta celda de código se encuentra los ejercicios 1, 2 y 3.

class Numero:
    
    def __init__(self, num): 
      if num < 1 or num > 4000:
        raise ValueError("No se soportan números menores de 1 ni mayores de 3999.")
      self.num = num
      self.rom = self.convertirRom(num)


    def convertirRom(self, num):
      m = ["", "M", "MM", "MMM"]
      c = ["", "C", "CC", "CCC", "CD", "D", "DC", "DCC", "DCCC", "CM"]
      x = ["", "X", "XX", "XXX", "XL", "L", "LX", "LXX", "LXXX", "XC"]
      i = ["", "I", "II", "III", "IV", "V", "VI", "VII", "VIII", "IX"]

      millares = m[num // 1000]
      centenas = c[(num % 1000) // 100]
      decenas = x[(num % 100) // 10]
      unidades = i[num % 10]
  
      return millares + centenas + decenas + unidades
    
    def convertirAra(self, rom):
      rom = rom.upper()
      self.is_romano(rom)
      romanos = {'I': 1, 'V': 5, 'X': 10, 'L': 50, 'C': 100, 'D': 500, 'M': 1000}
      total = 0
      prev = 0
      for letra in rom[::-1]:
        valor = romanos[letra]
        total += valor if valor >= prev else -valor
        prev = valor
      
      return total
    
    def suma_romano(self, rom):
      if type(rom) is not str:
        raise ValueError("El número Romano debe de ser una cadena de caracteres.")
      num2 = self.convertirAra(rom.upper())
      if self.num + num2 > 3999:
        raise ValueError('El resultado de la suma no puede ser mayor de 3999 en número romano.')
      self.num += num2
      self.rom = self.convertirRom(self.num)

      return


    def imprime(self):
      print("Número Arábigo: ", self.num, " Número Romano: ", self.rom)

      return
 
    def is_romano(self, rom):
      if type(rom) is not str:
        raise ValueError("El número Romano debe de ser una cadena de caracteres.")
      rom = rom.upper()

      for c in rom:
        if c not in ('M', 'C', 'D', 'X', 'L', 'I', 'V'):
          raise ValueError("El número romano tiene carácteres no válidos.")
      
      #Nota: existe una infinidad de controles que se pueden aplicar sobre la forma incorrecta de un número romano. 
      #En este ejercicio sólo se aplicará este control.
      if 'MMMM' in rom or 'CCCC' in rom or 'DDDD' in rom or 'XXXX' in rom or 'LLLL' in rom or 'IIII' in rom or 'VVVV' in rom:
        raise ValueError("El número romano no es válido.")
      
      return True

try:
  x = Numero(1672)
  x.imprime()
  x.suma_romano("xxx")
  x.imprime()
except ValueError as err:
  print("Error: ", err)
```

En la clase MejorNumero, crea un nuevo método llamado iterar() que reciba una lista con 3 números romanos. A continuación, iterando sobre los elementos de la lista llamará a la función suma_romano(). **NOTA:** el mensaje de error ha de seguir el siguiente formato --> print('Ha fallado el número', 5). 

```python
# En esta celda de código se encuentra los ejercicios 4 y 5.

class MejorNumero(Numero):

  def __init__(self, num): 
    super().__init__(num)

  def resta(self, rom):
    n2 = super().convertirAra(rom)
    if (self.num - n2) < 1:
      raise ValueError('El resultado de la resta no puede ser cero ni negativa en número romano.')
    self.num -= n2
    self.rom = super().convertirRom(self.num)

    return
  
  def multiplicacion(self, rom):
    n2 = super().convertirAra(rom)
    if (self.num * n2) > 3999:
      raise ValueError('El resultado de la multiplicación no puede ser mayor de 3999 en número romano.')
    self.num *= n2
    self.rom = super().convertirRom(self.num)

    return
  
  def iterar(self, r1, r2, r3):
    l = []
    l.append(r1)
    l.append(r2)
    l.append(r3)

    for n in l:
      try: 
        self.suma_romano(n)
      except ValueError as err:
        print('Error: ', err)
        try:
          print('Ha fallado el número: ', self.convertirAra(n))
        except:
          print('Ha fallado el número: ', n)

    return 

try: 
  x = MejorNumero(8)
  x.imprime()
  x.resta('i')
  x.imprime()

  x.multiplicacion('II')
  x.imprime()

  x.iterar('iii', 'xxix', 'i')
  x.imprime()
except ValueError as err:
  print('Error: ', err)
```
